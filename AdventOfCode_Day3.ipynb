```
def insert_brackets(string, index):
    return string[:index] + "[" + string[index] + "]" + string[index+1:]

def make_move(row_index, col_index, row_move = 1, col_move = 3):
    col_index += col_move
    row_index += row_move
    return row_index, col_index

def expand_map(data):
    return [data[i]*100 for i in range(len(data))]
def count_trees(data, row_move =1, col_move =3):
    counter = 0
    col_index = 0
    row_index = 0
    while row_index<len(data):
        if data[row_index][col_index]=="#":
            counter+=1
        row_index, col_index = make_move(row_index, col_index, row_move, col_move)
    return counter
```

## Task 2

> Determine the number of trees you would encounter if, for each of the following slopes, you start at the top-left corner and traverse the map all the way to the bottom:
>
>Right 1, down 1.

>Right 3, down 1. (This is the slope you already checked.)

>Right 5, down 1.

>Right 7, down 1.

>Right 1, down 2.

>In the above example, these slopes would find 2, 7, 3, 4, and 2 tree(s) respectively; multiplied together, these produce the answer 336.
>
>What do you get if you multiply together the number of trees encountered on each of the listed slopes?
>

```
slopes = [(1,1), (3,1), (5,1), (7,1), (1,2)]
```

```
result = 1
for slope in slopes:
    col_move, row_move = slope
    result *= count_trees(expand_map(data), row_move=row_move,col_move =col_move)

print(result)
```
